```python
def get_age_range():
    """
    Finds the minimum and maximum age of the patients in the dataset.
    Returns:
    tuple: A tuple containing the minimum and maximum age.
    """
    # Get the minimum and maximum values from the 'age' column
    min_age = df['age'].min()
    max_age = df['age'].max()
    return min_age, max_age
# Get the minimum and maximum age of patients
min_age, max_age = get_age_range()
print("Minimum age:", min_age)
print("Maximum age:", max_age)
```

```python
total_cost_by_treatment = df.groupby('treatment_type')['cost'].sum()
print(total_cost_by_treatment)
```

```python
 count=df.groupby(['treatment_type', 'gender']).size()
print(count)    #make count t
```

```python
outcome_counts = df.groupby(['treatment_type', 'treatment_outcome']).size().unstack(fill_value=0)
print(outcome_counts)
```

```python
pivot_table = df.pivot_table(values='cost', index='treatment_type', columns='treatment_outcome', aggfunc='sum', fill_value=0)
print(pivot_table)
```

```python
import matplotlib.pyplot as plt

outcome_counts.plot(kind='bar', stacked=True)
plt.title('Treatment Outcomes by Treatment Type')
plt.xlabel('Treatment Type')
plt.ylabel('Number of Patients')
plt.legend(title='Treatment Outcome')
plt.show()
```

```python
average_scores_by_outcome = df.groupby('treatment_outcome').agg({
    'air_pollution': 'mean',
    'dust_allergy': 'mean',
    'obesity': 'mean'
})
print(average_scores_by_outcome)
```
